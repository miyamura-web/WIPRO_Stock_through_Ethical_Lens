## WIPRO's Stock through Ethical Lens

```python
import yfinance as yf
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')

stock_data = yf.download('WIPRO.NS', start = '2015-01-01')
```

```python
# Keep only columns related to 'WIPRO.NS'
stock_data = stock_data.xs('WIPRO.NS', axis=1, level=1)     # This drops the 'Ticker' level, keeping only 'Close', 'High', etc.
stock_data.head()
```

### Data of last 10 years with Ethical issues

```python
import matplotlib.pyplot as plt
import pandas as pd

stock_data['Close'].plot(figsize=(12,6))

# Add annotations
events = {
    '2019-05-10': 'Whistleblower complaint',
    '2021-06-15': 'Ethical recognition award',
    '2022-11-14': 'Leadership Ethics Training',
    '2023-02-20': 'Sustainability report released',
}

for date, label in events.items():
    plt.axvline(pd.to_datetime(date), color='red' if "complaint" in label else 'green', linestyle='--', linewidth=1)
    plt.text(pd.to_datetime(date), stock_data['Close'].max()*0.95, label, rotation=45, ha='left', va='top', fontsize=9, color='black')

plt.title("Wipro Stock Price vs Ethics-Related Events")
plt.xlabel("Date")
plt.ylabel("Closing Price (INR)")
plt.grid(True)
plt.show()
```

#### 1. Does unethical things affect stock price ?

```python

stock_data = yf.download('WIPRO.NS', start='2019-01-01', end='2021-01-01')

# Plot the stock price
plt.figure(figsize=(14, 7))
plt.plot(stock_data.index, stock_data['Close'], label='WIPRO Stock Price', color='blue')
plt.title('Wipro Stock Price vs Ethics-Related Events')
plt.xlabel('Date')
plt.ylabel('Closing Price (INR)')
plt.grid(True)

# Ethics-related events
events = {
    '2019-11-15': 'Whistleblower Complaint'
}

# Annotate each event with a box and arrow
for date_str, label in events.items():
    date = pd.to_datetime(date_str)
    price = stock_data.loc[date, 'Close'] if date in stock_data.index else stock_data['Close'].loc[stock_data.index.get_loc(date, method='nearest')]
    
    plt.annotate(label,
                 xy=(date, price),
                 xytext=(date, price + 40),
                 textcoords='data',
                 fontsize=14,
                 bbox=dict(boxstyle="round,pad=0.3", edgecolor='black', facecolor='lightyellow'),
                 arrowprops=dict(arrowstyle="->", color='gray'))

plt.legend()
plt.tight_layout()
plt.show()
```

#### Comparison with Volkswagon

```python

stock_data_vk = yf.download('VWAGY', start='2015-01-01', end='2016-04-01')

# Plot the stock price
plt.figure(figsize=(14, 7))
plt.plot(stock_data_vk.index, stock_data_vk['Close'], label='VOLKSWAGON Stock Price', color='blue')
plt.title('Volkswagon Stock Price vs Ethics-Related Events')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.grid(True)


# Ethics-related events
events = {
   '2015-09-16': 'Whistleblower Complaint'
}

# Annotate each event with a box and arrow
for date_str, label in events.items():
    date = pd.to_datetime(date_str)
    price = stock_data_vk.loc[date, 'Close'] if date in stock_data_vk.index else stock_data_vk['Close'].loc[stock_data_vk.index.get_loc(date, method='nearest')]
    
    plt.annotate(label,
                 xy=(date, price),
                 xytext=(date, price + 5),
                 textcoords='data',
                 fontsize=14,
                 bbox=dict(boxstyle="round,pad=0.3", edgecolor='black', facecolor='lightyellow'),
                 arrowprops=dict(arrowstyle="->", color='gray'))

plt.legend()
plt.tight_layout()
plt.show()
```
